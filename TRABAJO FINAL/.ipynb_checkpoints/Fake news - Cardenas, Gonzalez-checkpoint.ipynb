```python
#Create a function to scrape the site
def scrape_website(page_number):
  page_num = str(page_number) #Convert the page number to a string
  URL = 'https://www.politifact.com/factchecks/list/?page='+page_num #append the page number to complete the URL
  webpage = requests.get(URL)  #Make a request to the website
  #time.sleep(3)
  soup = BeautifulSoup(webpage.text, "html.parser") #Parse the text from the website
  #Get the tags and it's class
  statement_footer =  soup.find_all('footer',attrs={'class':'m-statement__footer'})  #Get the tag and it's class
  statement_quote = soup.find_all('div', attrs={'class':'m-statement__quote'}) #Get the tag and it's class
  statement_meta = soup.find_all('div', attrs={'class':'m-statement__meta'})#Get the tag and it's class
  target = soup.find_all('div', attrs={'class':'m-statement__meter'}) #Get the tag and it's class
  #loop through the footer class m-statement__footer to get the date and author
  for i in statement_footer:
    link1 = i.text.strip()
    name_and_date = link1.split()
    first_name = name_and_date[1]
    last_name = name_and_date[2]
    full_name = first_name+' '+last_name
    month = name_and_date[4]
    day = name_and_date[5]
    year = name_and_date[6]
    date = month+' '+day+' '+year
    dates.append(date)
    authors.append(full_name)
  #Loop through the div m-statement__quote to get the link
  for i in statement_quote:
    link2 = i.find_all('a')
    statements.append(link2[0].text.strip())
 #Loop through the div m-statement__meta to get the source
  for i in statement_meta:
    link3 = i.find_all('a') #Source
    source_text = link3[0].text.strip()
    sources.append(source_text)
  #Loop through the target or the div m-statement__meter to get the facts about the statement (True or False)
  for i in target:
    fact = i.find('div', attrs={'class':'c-image'}).find('img').get('alt')
    targets.append(fact)
```

```python
#Creacion de listas para el web scrapping
authors = []
dates = []
statements = []
sources = []
targets = []
```

```python
#Loop through 'n-1' webpages to scrape the data
n=65
for i in range(1, n):
    scrape_website(i)
```

<h2><strong>Creacion del nuevo dataframe:</strong></h2>

```python
#Create a new dataFrame 
data = pd.DataFrame(columns = ['author',  'statement', 'source', 'date', 'target']) 
data['author'] = authors
data['statement'] = statements
data['source'] = sources
data['date'] = dates
data['target'] = targets
#Show the data set
data
```

```python
#Función para obtener un número binario del target
def getBinaryNumTarget(text):
    if text == 'true':
        return 1
    else:
        return 0
```

```python
#Función para obtener solo valores verdaderos o falsos del target
def getBinaryTarget(text):
    if text == 'true':
        return 'REAL'
    else:
        return 'FAKE'
```

```python
data['BinaryTarget'] = data['target'].apply(getBinaryTarget)
data['BinaryNumTarget'] = data['target'].apply(getBinaryNumTarget)
data
```

# Limpieza de data

En los datos de texto, a menudo existen componentes que pueden no ser de ayuda para nosotros. El objetivo principal de esta sección será eliminar esos componentes, de modo que podamos tener un análisis más fluido. Planeamos crear una función aquí que será responsable de eliminar cualquier palabra innecesaria (palabras vacías) de los datos proporcionados, y utilizaremos esta función cuando sea necesario durante nuestro análisis y predicción.

```python
data.isnull().sum()
```

```python
stop_words = stopwords.words('english')
stop_words.extend(['from', 'subject', 're', 'edu', 'use'])
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 2 and token not in stop_words:
            result.append(token)
            
    return result
```

<h2><strong>Análisis de sentimientos del contenido de la noticia:</strong></h2>

```python
#Sentimientos
from textblob import TextBlob

bloblist_title = list()

sentimiento=news['clean_joined_text']
for row in sentimiento:
    blob = TextBlob(row)
    bloblist_title.append((row,blob.sentiment.polarity, blob.sentiment.subjectivity))
    df_usa_polarity_title = pd.DataFrame(bloblist_title, columns = ['sentence','sentiment','polarity'])
 
    def f_tags(df_usa_polarity_title):
        if df_usa_polarity_title['sentiment'] > 0:
            val = "Positive"
        elif df_usa_polarity_title['sentiment'] == 0:
            val = "Neutral"
        else:
            val = "Negative"
        return val

df_usa_polarity_title['Sentiment_Type'] = df_usa_polarity_title.apply(f_tags, axis=1)

plt.figure(figsize=(10,10))
sns.set_style("whitegrid")
ax = sns.countplot(x="Sentiment_Type", data=df_usa_polarity_title)
```

```python
#Distribución de palabras en el contenido *** cuantas palabras hay en el contenido de la noticia
maxlen = -1
for doc in news.clean_joined_text:
    tokens = nltk.word_tokenize(doc)
    if(maxlen<len(tokens)):
        maxlen = len(tokens)
print("El numero máximo de palabras que tiene el contenido es de =", maxlen)
fig = px.histogram(x = [len(nltk.word_tokenize(x)) for x in news.clean_joined_text], nbins = 50)
fig.show()
```

```python
#Entrenamiento
X_train, X_test, y_train, y_test = train_test_split(news.clean_joined_text, news.BinaryNumTarget, test_size = 0.2,random_state=2)
vec_train = CountVectorizer().fit(X_train)
X_vec_train = vec_train.transform(X_train)
X_vec_test = vec_train.transform(X_test)
```

```python
#Modelo:
model = LogisticRegression(C=2.5)
model.fit(X_vec_train, y_train)
predicted_value = model.predict(X_vec_test)
accuracy_value = roc_auc_score(y_test, predicted_value)
print('Acurracy',accuracy_value)
```

<h2><strong>Implementación de matriz de confusión:</strong></h2><br>

```python
prediction = []
for i in range(len(predicted_value)):
    if predicted_value[i].item() > 0.5:
        prediction.append(1)
    else:
        prediction.append(0)
matriz = confusion_matrix(list(y_test), prediction)
plt.figure(figsize = (10, 10))
sns.heatmap(matriz, annot = True,fmt='g')
```
