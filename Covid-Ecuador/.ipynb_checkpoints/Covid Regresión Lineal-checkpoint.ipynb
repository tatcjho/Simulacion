```python
import pandas as pd
df=pd.read_csv('dataset.csv', sep = ';')
```

```python
df.head()
```

```python
from datetime import datetime,timedelta
import datetime as dt
FMT = '%d/%m/%Y'
df['date'] = df['date'].map(lambda x : (datetime.strptime(x, FMT) - datetime.strptime("01/01/2020", FMT)).days)

#df['date'] = [dt.datetime.strptime(x,'%d/%m/%Y') for x in df['date']] 
```

```python
df['date']=df['date'].map(dt.datetime.toordinal)
```

```python
df.head()
```

```python
df=df[['date','total_cases']]
df
```

```python
df.describe()
```

```python
import numpy as np  
import matplotlib.pyplot as plt  

plt.figure(figsize = (10,5))
x = df['date'].values.reshape(-1,1)
y = df['total_cases'].values.reshape(-1,1)
plt.plot(x, y,'.', markersize = 5, label ='')
plt.title('Casos')  
plt.xlabel('Fecha')  
plt.ylabel('Contagios') 

plt.show()
```

```python
#X = df.iloc[:, :-1].values  
#y = df.iloc[:, 1].values
x = df['date'].values.reshape(-1,1)
y = df['total_cases'].values.reshape(-1,1)
```

División del dataset en datos para el entrenamiento y el test

```python
from sklearn.model_selection import train_test_split  
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42) 
```

Entrenamiento del modelo de Regresion Lineal

```python
#Training: Regresion Lineal
from sklearn.linear_model import LinearRegression  
regressor = LinearRegression()  
regressor.fit(X_train, y_train) 
```

```python
#Making Predictions: Regresion Lineal
y_pred = regressor.predict(X_test)  
#y_pred es una matriz numpy que contiene todos los valores predichos para los valores de entrada en la serie X_test
#Comparacion manual
df3 = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()}) 
#Aunque nuestro modelo no es muy preciso, los porcentajes predichos son cercanos a los reales.
df3 
```

```python
from datetime import datetime

date_string = "3/11/2021"
hoy = "15/4/2021"
FMT = '%d/%m/%Y'
nuevaFecha =  (datetime.strptime(date_string, FMT) - datetime.strptime("01/01/2020", FMT)).days
hoy1 =  (datetime.strptime(hoy, FMT) - datetime.strptime("01/01/2020", FMT)).days
y_pred = regressor.predict(np.array([[nuevaFecha]]))
df3 = pd.DataFrame({"Hoy": df[df.date == hoy1]['total_cases'], date_string: y_pred[0].round()}) 
```
